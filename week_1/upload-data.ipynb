```python
import pandas as pd
pd.__version__
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
# run this cell when the Postgres Docker container is running
engine.connect()
```

```python
df_green = pd.read_csv('green_tripdata_2019-01.csv.gz')
df_green.shape
```

```python
df_green.lpep_pickup_datetime = pd.to_datetime(df_green.lpep_pickup_datetime)
df_green.lpep_dropoff_datetime = pd.to_datetime(df_green.lpep_dropoff_datetime)
```

```python
print(pd.io.sql.get_schema(df_green, name='green_taxi_trip_data', con=engine))
```

```python
df_green.to_sql(name='green_taxi_trip_data', con=engine, chunksize=1000, if_exists='append')
```

```python
df_zone = pd.read_csv('taxi+_zone_lookup.csv')
df_zone
```

```python
df_zone.to_sql(name='taxi_zones', con=engine, chunksize=1000, if_exists='append')
```
